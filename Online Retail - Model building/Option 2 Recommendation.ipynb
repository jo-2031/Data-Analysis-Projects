```python
#importing the libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Upload the dataset using pandas
df = pd.read_csv("Online Retail.csv")
df
```

# Data Cleaning 

```python
df.shape
```

```python
df = df.loc[df['Quantity'] > 0]
```

```python
df.info()
```

```python
# Delete the null used the Customer ID
df['CustomerID'].isna().sum()
df = df.dropna(subset=['CustomerID'])
```

```python
# Using pivot table describe the stockcode with customer ID
c_item = df.pivot_table(
    index='CustomerID',
    columns='StockCode',
    values='Quantity',
    aggfunc='sum'
)
c_item.loc[12481:].head()
```

```python
#Printing the shape (size of the dataset after the data Cleaning)
print(c_item.shape)
c_item = c_item.applymap(lambda x: 1 if x > 0 else 0)
```

# Collaborative Filtering

```python
#Determine the cosine similarity from the data
df2 = pd.DataFrame(cosine_similarity(c_item))
df2.head()
```
